Start with simple MNIST example

Conv nets take shapes of 

```python
(image_height, image_width, image_channels)
```

does not including the batch dimension

````python
from keras import layers 
from keras import models
from keras.datasets import mnist
from keras.utils import to_categorical
from keras import optimizers

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1))) 
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu')) 
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Flatten 3D tensor into 1D
model.add(layers.Flatten())
# Reshape
model.add(layers.Dense(64, activation='relu'))
# Classify
model.add(layers.Dense(10, activation='softmax'))

````

````python
model.summary()
````

````python
# import data

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# reshape the data to what network expects
train_images = train_images.reshape((60000, 28, 28, 1))
# normalize
train_images = train_images.astype('float32') / 255
# reshape
test_images = test_images.reshape((10000, 28, 28, 1))
# normalize
test_images = test_images.astype('float32') / 255
# vectorize
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5, batch_size=64)
````

````python
score, acc = model.evaluate(test_images, test_labels)
print('Test score:', score)
print('Test accuracy:', acc)
````

Wow! CNNs are really powerful. We got >99% accuracy on the test data. Holy mackerel.

## Conv Nets explained

### Convolutions

A convolution works by 

Step 1) sliding a window (usually 3x3) over the 3D input feature map, stopping at every location, and extracting the 3D patch of surrounding features.

Step 2) Transform 3D patch of output features into 1D vector by doing a tensor product between the 3D patch and the weights of the layer (convolution kernel).

Step 3) Reassemble all of these 1D vectors into 3D output map.

Note: Typically the output map has the same spatial dimensions as input with some exceptions.

![Image](https://s3-us-west-2.amazonaws.com/mishalaskin/dl/convolution.png)


Conv nets take a 3D tensor (height, width, channel) where the channel is the pixel value (scalar if it's black and white, and a 3 unit vector if it's RGB). 

A convolution takes a 3D tensor, extracts patches from it and applies a transformation to all of those patches.

The first convolution inputs this 3D tensors and outputs a 3D tensor. The new tensor also has a height and width as well as a channel. However, the channel is no longer the RGB or grayscale pixel value, it's now a filter. Filters store info about discovered features (e.g. edges, circles, etc) and the new width / height image show you how prevalent this feature is in the image. 

1. Filters (or features) are stored in the channel axis
2. The 2D image stored for each filter shows where a filter occurs in the image

![Image](https://s3-us-west-2.amazonaws.com/mishalaskin/dl/convnetfilter.png)


Convolutions are defined by two key parameters:

a. Size of the patches extracted from the inputs—These are typically 3 × 3 or 5 × 5. In the example, they were 3 × 3, which is a common choice.

b. Depth of the output feature map—The number of filters computed by the convolution. The example started with a depth of 32 and ended with a depth of 64.

In keras, you implement one with

```python
Conv2D(number_of_filters, (patch_height, patch_width)).
```

#### Border effects and padding 

If a X by X feature map cannot accommodate X x X convolutions (example 5 x 5 feature map, only accommodates 9 3x3 convolutions), then you need to add padding (zero values) to the border of the feature map.

#### Strides

Usually set to 1, but you can make the larger even though it's not common. MaxPooling is preferred.

### MaxPooling

MaxPooling reduces the image size by scanning the image with a (typically 2X2 with stride 2) patch and taking the maximum value from the 2X2 matrix. This is done for 2 reasons:

1. Compressing the image allows a convolution filter to learn abstract features
2. it dramatically reduces the number of parameters in the model

````python
import os, shutil

# set path ro original directory

original_training_dir  = '../data/catsdogs/train'

base_dir = './cats_and_dogs_small'
os.mkdir(base_dir)

# train / val / test dirs

train_dir = os.path.join(base_dir,'train')
os.mkdir(train_dir)

validation_dir = os.path.join(base_dir,'validation')
os.mkdir(validation_dir)

test_dir = os.path.join(base_dir,'test')
os.mkdir(test_dir)

# train cats / dogs dirs

train_cats_dir = os.path.join(train_dir,'cats')
os.mkdir(train_cats_dir)

train_dogs_dir = os.path.join(train_dir,'dogs')
os.mkdir(train_dogs_dir)

# val cats / dogs dirs

validation_cats_dir = os.path.join(validation_dir,'cats')
os.mkdir(validation_cats_dir)

validation_dogs_dir = os.path.join(validation_dir,'dogs')
os.mkdir(validation_dogs_dir)

# test cats / dogs dirs

test_cats_dir = os.path.join(test_dir,'cats')
os.mkdir(test_cats_dir)

test_dogs_dir = os.path.join(test_dir,'dogs')
os.mkdir(test_dogs_dir)
````

````python
# copy files from original data source
# 1000 training, 500 validation, 500 test

# cats
fnames = ['cat.{}.jpg'.format(i) for i in range(1000)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(train_cats_dir, fname)
    shutil.copyfile(src,dst)
    
fnames = ['cat.{}.jpg'.format(i) for i in range(1000,1500)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(validation_cats_dir, fname)
    shutil.copyfile(src,dst)

fnames = ['cat.{}.jpg'.format(i) for i in range(1500,2000)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(test_cats_dir, fname)
    shutil.copyfile(src,dst)

# dogs 

fnames = ['dog.{}.jpg'.format(i) for i in range(1000)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(train_dogs_dir, fname)
    shutil.copyfile(src,dst)
    
fnames = ['dog.{}.jpg'.format(i) for i in range(1000,1500)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(validation_dogs_dir, fname)
    shutil.copyfile(src,dst)

fnames = ['dog.{}.jpg'.format(i) for i in range(1500,2000)]
for fname in fnames:
    src = os.path.join(original_training_dir,fname)
    dst = os.path.join(test_dogs_dir, fname)
    shutil.copyfile(src,dst)
````

## The actual model

The images are larger and more complex than in the MNIST example. For this reason, we'll make the network bigger and add another Conv2D layer, which will alse help reduce the feature map before the final Flatten step.



````python
model = models.Sequential()

# 32 filters, 3X3 windows
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
# max pool 2X2 window
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Flatten())
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

#1. Optimizer - RMSprop
#2. Loss - binary cross entr
model.compile(loss='binary_crossentropy',optimizer=optimizers.RMSprop(lr=1e-4),metrics=['acc'])
````

````python
model.summary()
````

## Reading in the data

1 Read the picture files.

2 Decode the JPEG content to RGB grids of pixels.

3 Convert these into floating-point tensors.

4 Rescale the pixel values (between 0 and 255) to the [0, 1] interval (as you know,
neural networks prefer to deal with small input values).

Keras has a module for doing these things

````python
from keras.preprocessing.image import ImageDataGenerator

# rescale by 1/255 (normalizes to [0,1])
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# resize all images to 150 x 150 (what network expects)
# since we're using binary cross entropy, we need binary labels, hence class_mode='binary'

train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size=(150,150),
                                                   batch_size=20,
                                                   class_mode='binary')

validation_generator=test_datagen.flow_from_directory(validation_dir,
                                                     target_size=(150,150),
                                                     batch_size=20,
                                                     class_mode='binary')
````

### understanding generators:

A generator is an iterator. Quick example 

```python
def generator(): i=0
    while True:
        i += 1
        yield i
for item in generator():
    print(item)
    if item > 4:
        break
```

which prints

```python
1 
2 
3 
4 
5
```

Let's look at what our generator prints. It should print 20 images with resolution 150x150 and a 3 component vector of RGB values

````python
for i, data in enumerate(train_generator):
    data_batch = data[0]
    labels_batch= data[1]
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    if i > 5:
        break
````

### Final notes about generators

Generators loop endlessly, so you need to tell them when to stop. For Keras, there is a ```fit_generator``` function, which fits generator data to the model. To tell Keras when to stop, you need to specify ```steps_per_epoch```. 

We have generators that batch 20 images. There are 2000 training images (cats + dogs) total, so we want ```steps_per_epoch = 100```

Finally, the validation data can be either a generator or a numpy tuple. If it's a generator, it needs to know when to stop. Again, we have generators that batch 20 images. There are 1000 validation images (cats + dogs) total, so we want ```validation_steps = 50```

````python
history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,
      epochs=5,
      validation_data=validation_generator,
      validation_steps=50)

model.save('cats_and_dogs_small_1.h5')
````

````python
import matplotlib.pyplot as plt
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
````

We kept it to 5 epochs for time efficiency. If we did more, say 20-40, we would find that the model overfits. So we need to regularize. we'll add

1. data augmentation
2. dropout


````python
# data augmentation generator
datagen = ImageDataGenerator(
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')


#1 rotation_range is a value in degrees (0–180), a range within which to ran- domly rotate pictures.
#2 width_shift and height_shift are ranges (as a fraction of total width or height) within which to randomly translate pictures vertically or horizontally.
#3 shear_range is for randomly applying shearing transformations.
#4 zoom_range is for randomly zooming inside pictures.
#5 horizontal_flip is for randomly flipping half the images horizontally—
#relevant when there are no assumptions of horizontal asymmetry (for example,
#real-world pictures).
#6 fill_mode is the strategy used for filling in newly created pixels, which can
appear after a rotation or a width/height shift.
````

````python
# look at some augmented images

# import module w image processing
from keras.preprocessing import image
fnames = [os.path.join(train_cats_dir, fname) for 
          fname in os.listdir(train_cats_dir)]

# choose image to augment
img_path = fnames[1]
# read and resize image
img = image.load_img(img_path, target_size=(150, 150))
#convert image to numpy array w shape (150,150,3)
x = image.img_to_array(img)
# reshape to (1,150,150,3)
x = x.reshape((1,) + x.shape)
# generates batch of augmented images
i=0
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i % 4 == 0:
        break
plt.show()
````

````python
# define network with dropout


model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
# note how we do dropout on dense layers, not conv layers
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
````

````python
# augment the data

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

# only augment training data, not validation data

test_datagen = ImageDataGenerator(rescale=1./255)

# resize images to 150 x 150
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary')

history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,
      epochs=5,
      validation_data=validation_generator,
      validation_steps=50)

model.save('cats_and_dogs_small_2.h5')
````
